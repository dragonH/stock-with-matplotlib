```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.read_csv('AAPL.csv', parse_dates = ['Date']).plot(x = 'Date', y = 'Close', ax = plt.gca(), label = 'AAPL')
pd.read_csv('MSFT.csv', parse_dates = ['Date']).plot(x = 'Date', y = 'Close', color = 'orange', ax = plt.gca(), label = 'MSFT')

```
